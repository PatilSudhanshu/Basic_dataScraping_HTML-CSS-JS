```python
def getCurrencyChange(price,fromCurrency,toCurrency):
    curl = 'https://api.coinbase.com/v2/exchange-rates'
    cheaders =  {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36'
    }
    param = {
        'currency' : fromCurrency
    }

    currencies = rq.get(url= curl, headers= cheaders, params=param)

    currencyData = currencies.json()['data']['rates']
    
    if toCurrency in currencyData:
        exchangeRate = currencyData[toCurrency]
    else:
        exchangeRate = None


    if exchangeRate:
        convertedPrice = price * float(exchangeRate)
        return convertedPrice ,exchangeRate
    else:
        print(f'Rate for {toCurrency} not found')
        return None, None
```
